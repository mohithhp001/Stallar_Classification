```python
df.drop('cam_col',1,inplace=True)
```

```python
df_feature=df.drop('class',1)
df_feature.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from  xgboost  import XGBClassifier
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
from sklearn.metrics import classification_report,accuracy_score
```

```python
df_feature.head()
```

```python
sns.heatmap(df_feature.corr(),annot=True)
```

```python
df.head()
```

```python
X=df_feature
Y=df['class']
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42,)
x_sm, y_sm = sm.fit_resample(X, Y)
```

```python
x_sm.shape,y_sm.shape
```

```python
df_scaled_feat=ss.fit_transform(x_sm)
```

```python
df_feature=pd.DataFrame(df_scaled_feat,columns=df_feature.columns)
```

```python
df_feature.head()
```

```python
x=df_feature
y=y_sm

xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=10)
print(xtrain.shape,ytrain.shape)
print(xtest.shape,ytest.shape)
```

# Logistic Regression

```python
lr=LogisticRegression()
lr.fit(xtrain,ytrain)
ypred_lr=lr.predict(xtest)
print(classification_report(ytest,ypred_lr))
```

```python
accuracy_lr=accuracy_score(ytest,ypred_lr)
accuracy_lr
```

# KNN


```python
knn=KNeighborsClassifier(n_neighbors=4)
knn.fit(xtrain,ytrain)
ypred_knn=knn.predict(xtest)
print(classification_report(ytest,ypred_knn))
accuracy_knn=accuracy_score(ytest,ypred_knn)
```

```python
ytrain_predict=knn.predict(xtrain)
Train_accuracy=accuracy_score(ytrain,ytrain_predict)
Train_accuracy
```

```python
ytest_predict=knn.predict(xtest)
Test_accuracy=accuracy_score(ytest,ytest_predict)
Test_accuracy
```
